```python
# Task 1 — Explore the Menu

menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

# part 1 - print menu grouped by category
# loop through each category first, then find matching items

categories = ["Starters", "Mains", "Desserts"]

for category in categories:
    print("===== " + category + " =====")

    for item in menu:
        if menu[item]["category"] == category:

            if menu[item]["available"] == True:
                status = "[Available]"
            else:
                status = "[Unavailable]"

            # formatting price to 2 decimal places
            price_str = "₹" + str(format(menu[item]["price"], ".2f"))
            print(f"{item:<16} {price_str:<9}  {status}")

    print()
```

```python
# part 2 - menu stats using dictionary methods

# total number of items
total_items = len(menu)
print("Total items on menu :", total_items)

# count available items
available_count = 0
for item in menu:
    if menu[item]["available"] == True:
        available_count = available_count + 1
print("Available items     :", available_count)

# find the most expensive item
expensive_name  = ""
expensive_price = 0

for item in menu:
    if menu[item]["price"] > expensive_price:
        expensive_price = menu[item]["price"]
        expensive_name  = item

print("Most expensive item : " + expensive_name + " (₹" + format(expensive_price, ".2f") + ")")
print()

# all items priced under 150
print("Items priced under ₹150:")
for item in menu:
    if menu[item]["price"] < 150:
        print("- " + item + "  ₹" + format(menu[item]["price"], ".2f"))
```

```python
# Task 2 — Cart Operations

cart = []

# ─────────────────────────────────────────
# add item logic
# ─────────────────────────────────────────

def add_item(item, qty):
    # check if item exists in menu
    if item not in menu:
        print(item + " not found in menu")
        return

    # check if item is available
    if menu[item]["available"] == False:
        print(item + " is currently unavailable")
        return

    # if item already in cart just update quantity
    already_in_cart = False
    for thing in cart:
        if thing["item"] == item:
            thing["quantity"] = thing["quantity"] + qty
            print(item + " quantity updated!")
            already_in_cart = True

    # otherwise add as new entry
    if already_in_cart == False:
        cart.append({"item": item, "quantity": qty, "price": menu[item]["price"]})
        print(item + " added to cart!")

# ─────────────────────────────────────────
# remove item logic
# ─────────────────────────────────────────

def remove_item(item):
    found = False
    for thing in cart:
        if thing["item"] == item:
            cart.remove(thing)
            print(item + " removed from cart!")
            found = True
            break
    if found == False:
        print(item + " not found in cart")

# ─────────────────────────────────────────
# update quantity logic
# ─────────────────────────────────────────

def update_qty(item, new_qty):
    found = False
    for thing in cart:
        if thing["item"] == item:
            thing["quantity"] = new_qty
            print(item + " quantity changed to " + str(new_qty))
            found = True
    if found == False:
        print(item + " not found in cart")

# ─────────────────────────────────────────
# simulate the order sequence
# ─────────────────────────────────────────

add_item("Paneer Tikka", 2)
print(cart)
print()

add_item("Gulab Jamun", 1)
print(cart)
print()

# this should update quantity to 3, not add a new entry
add_item("Paneer Tikka", 1)
print(cart)
print()

# this does not exist in menu
add_item("Mystery Burger", 1)
print(cart)
print()

# this exists but is marked unavailable
add_item("Chicken Wings", 1)
print(cart)
print()

remove_item("Gulab Jamun")
print(cart)
print()
```

```python
# final order summary

print("========== Order Summary ==========")

subtotal = 0

for thing in cart:
    line_total = thing["price"] * thing["quantity"]
    subtotal   = subtotal + line_total
    print(f"{thing['item']:<18} x{thing['quantity']:<4} ₹{line_total:.2f}")

gst   = round(subtotal * 5 / 100, 2)
total = round(subtotal + gst, 2)

print("------------------------------------")
print(f"Subtotal:                ₹{subtotal:.2f}")
print(f"GST (5%):                ₹ {gst:.2f}")
print(f"Total Payable:           ₹{total:.2f}")
print("====================================")
```

```python
# Task 3 — Inventory Tracker with Deep Copy

import copy

inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

# step 1 - deep copy inventory before making any changes
inventory_backup = copy.deepcopy(inventory)

# change one value to prove deep copy works
inventory["Paneer Tikka"]["stock"] = 999

print("inventory Paneer Tikka stock  :", inventory["Paneer Tikka"]["stock"])
print("backup Paneer Tikka stock     :", inventory_backup["Paneer Tikka"]["stock"])
print("deep copy is working — backup is safe!")
print()

# restore inventory back to original before continuing
inventory["Paneer Tikka"]["stock"] = 10
print("inventory restored back to original")
print()
```

```python
# step 2 - deduct cart quantities from inventory

for thing in cart:
    item = thing["item"]
    qty  = thing["quantity"]

    if inventory[item]["stock"] >= qty:
        inventory[item]["stock"] = inventory[item]["stock"] - qty
        print("deducted " + str(qty) + " from " + item)
    else:
        # not enough stock, deduct only what is available
        print("⚠ not enough stock for " + item + ", deducting only what is available")
        inventory[item]["stock"] = 0

print()

# step 3 - print reorder alerts for items at or below reorder level
for item in inventory:
    stock   = inventory[item]["stock"]
    reorder = inventory[item]["reorder_level"]

    if stock <= reorder:
        print("⚠ Reorder Alert: " + item + " — Only " + str(stock) + " unit(s) left (reorder level: " + str(reorder) + ")")

print()

# step 4 - print both to confirm they are different
print("current inventory:")
for item in inventory:
    print(f"{item:<15}: {inventory[item]['stock']}")

print()

print("backup inventory (unchanged):")
for item in inventory_backup:
    print(f"{item:<15}: {inventory_backup[item]['stock']}")
```

```python
# Task 4 — Daily Sales Log Analysis

sales_log = {
    "2025-01-01": [
        {"order_id": 1,  "items": ["Paneer Tikka", "Garlic Naan"],          "total": 220.0},
        {"order_id": 2,  "items": ["Gulab Jamun", "Veg Soup"],              "total": 210.0},
        {"order_id": 3,  "items": ["Butter Chicken", "Garlic Naan"],        "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4,  "items": ["Dal Tadka", "Garlic Naan"],             "total": 220.0},
        {"order_id": 5,  "items": ["Veg Biryani", "Gulab Jamun"],           "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6,  "items": ["Paneer Tikka", "Rasgulla"],             "total": 260.0},
        {"order_id": 7,  "items": ["Butter Chicken", "Veg Biryani"],        "total": 570.0},
        {"order_id": 8,  "items": ["Garlic Naan", "Gulab Jamun"],           "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9,  "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"],          "total": 270.0},
    ],
}

# step 1 - total revenue per day
print("Revenue per day:")
for date in sales_log:
    day_total = 0
    for order in sales_log[date]:
        day_total = day_total + order["total"]
    print(date + " : ₹" + str(day_total))

print()

# step 2 - best selling day
best_day     = ""
best_revenue = 0

for date in sales_log:
    day_total = 0
    for order in sales_log[date]:
        day_total = day_total + order["total"]
    if day_total > best_revenue:
        best_revenue = day_total
        best_day     = date

print("Best selling day: " + best_day + "  ₹" + str(best_revenue))
print()

# step 3 - most ordered item
# count how many orders each item appears in
item_count = {}

for date in sales_log:
    for order in sales_log[date]:
        for item in order["items"]:
            if item not in item_count:
                item_count[item] = 0
            item_count[item] = item_count[item] + 1

top_item  = ""
top_count = 0

for item in item_count:
    if item_count[item] > top_count:
        top_count = item_count[item]
        top_item  = item

print("Most ordered item: " + top_item + " (" + str(top_count) + " orders)")
print()
```

```python
# step 4 - add new day and reprint stats

sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

# reprint revenue per day with new day included
print("Updated revenue per day:")

best_day     = ""
best_revenue = 0

for date in sales_log:
    day_total = 0
    for order in sales_log[date]:
        day_total = day_total + order["total"]
    print(date + " : ₹" + str(day_total))

    if day_total > best_revenue:
        best_revenue = day_total
        best_day     = date

print()
print("New best selling day: " + best_day + "  ₹" + str(best_revenue))
print()
```

```python
# step 5 - numbered list of all orders using enumerate

print("All orders:")
print()

# collect all orders into one flat list first
all_orders = []
for date in sales_log:
    for order in sales_log[date]:
        all_orders.append((date, order))

# print each order numbered using enumerate
for number, (date, order) in enumerate(all_orders, start=1):
    items_str = ", ".join(order["items"])
    print(str(number) + ". [" + date + "] Order #" + str(order["order_id"]) + "  — ₹" + str(order["total"]) + " — Items: " + items_str)
```
